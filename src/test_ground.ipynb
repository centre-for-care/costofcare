```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as mtick
mpl.rc('font', family='Liberation Sans')
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./exec_times_toy.csv', index_col=0)
df.columns = ['Donor Pool Size', 'Exec Time']
```

```python
df.head(10)
```

```python
def get_y(x):
    b = 0.0099
    c = 0.11
    return c*(2**(b*x))
```

```python
df['$O(2^n)$'] = df['Donor Pool Size'].apply(get_y)
```

```python
ax = df.plot(x='Donor Pool Size', y=['Exec Time', '$O(2^n)$'], color=['#CF202A', '#41558c'], figsize=(9, 6))
ax.grid(linestyle='--', color='k', alpha=0.1, zorder=0)
ax.yaxis.set_major_locator(plt.MaxNLocator(6))
ax.set_ylabel('Time (seconds)')
plt.tight_layout()
sns.despine() 
```
